```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the data
url = "https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-07-07/coffee_ratings.csv"
df = pd.read_csv(url)

# 2. Save the raw data to your local Colab environment
df.to_csv('coffee_data.csv', index=False)
print("File saved as coffee_data.csv")

# 3. Process the data (Basic Cleaning)
# Let's filter for the main columns and remove missing values
cleaned_df = df[['country_of_origin', 'total_cup_points', 'variety', 'aroma', 'flavor']].dropna()

# 4. Visualize
plt.figure(figsize=(12, 6))
top_countries = cleaned_df.groupby('country_of_origin')['total_cup_points'].mean().sort_values(ascending=False).head(10)
top_countries.plot(kind='bar', color='brown')

plt.title('Top 10 Countries by Average Coffee Rating')
plt.xlabel('Country')
plt.ylabel('Average Total Cup Points')
plt.xticks(rotation=45)
plt.show()
```
